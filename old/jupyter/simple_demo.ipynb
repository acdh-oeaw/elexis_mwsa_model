```python
def categorize_by_label(df):
    relation_labels = df['relation'].unique()
    smallest_by_label = {}
    for relation in relation_labels:
        smallest_by_label[relation] = df[has_label(df, relation)]

    return smallest_by_label

def has_label(df, label):
    return df['relation'] == label


none = data[has_label(data, 'none') == True]
second_biggest = data.groupby('relation').count().word.sort_values(ascending=False)[1]
```

```python
def __calculate_max_similarity(target, spacy_doc):
        similarities = []
        for token in spacy_doc:
            if token.is_stop is False:
                similarities.append(target.similarity(token))
            #else:
            #print(token.text)
        if len(similarities) == 0:
            return 0.0
        return max(similarities)

def extract(data, feats):
    result = data.apply(lambda row: __calculate_max_similarity(row['word_processed'][0], row['processed_1']), axis=1)
    result2 = data.apply(lambda row:__calculate_max_similarity(row['word_processed'][0], row['processed_2']), axis=1)
    feats['simdiff_target'] = result-result2
    
extract(data, data)
```
